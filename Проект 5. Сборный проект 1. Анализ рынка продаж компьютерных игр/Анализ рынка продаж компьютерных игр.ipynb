### Шаг 3.1.  Посмотрите, сколько игр выпускалось в разные годы. Важны ли данные за все периоды?

````python
fig, ax = plt.subplots()
(data.groupby('year_of_release')['name'].count()
 .plot(title='Количество выпущенных игр в разные годы',
      grid=True,
      figsize=(10,5),
      style='o-'));

# Добавляем подписи к осям:
ax.set_xlabel('Год релиза');
ax.set_ylabel('Количество игр');
````

Агрегация по count() покажет нам число игр, выпущенных на всех платформах, в разрезе лет. Т.е. если одна и та же игра (name) вышла в некоем году на одной платформе, агрегат == 1. Если на 2-х платформах, агрегат == 2, ну и так далее.

Чтобы получить истинное количество «игр по годам», нам следует поискать дубликаты по (Годам - Названиям игр) и затем исключить из полного датафрейма все дубли за исключением одной, «первой», строки.

````python
data.duplicated(subset=['name','platform']).sum() # проверяем дубликаты по столбцам 'name' и 'platform'.
````

````python
data_without_duplicates = data.drop_duplicates(subset=['name','platform'], keep='first')  # удаляем данные дубликаты
````

````python
data_without_duplicates.duplicated(subset=['name','platform']).sum() # проверяем, остались ли дубликаты по столбцам 'name'
# и 'platform'
````

Теперь построим график на основании датафрейма data_without_duplicates, в котором нет дубликатов одновременно по столбцам 'name' и 'platform'. В таком случае если игра вышла в один год на разных платформах, то ей будет соответствовать только одна строка в датафрейме data_without_duplicates.

````python
fig, ax = plt.subplots()
(data_without_duplicates.groupby('year_of_release')['name'].count()
 .plot(title='Количество выпущенных игр в разные годы',
      grid=True,
      figsize=(10,5),
      style='o-'));

#  Добавляем подписи к осям:
ax.set_xlabel('Год релиза');
ax.set_ylabel('Количество игр');
````

---

Видим, что до 1995 года игр выпускалось меньше 200 штук. В 1995 году произошло существенное увеличение количества выпускаемых игр. Данные за период до 1995 года не важны.  
До 2008-2009 годов наблюдался тренд, связанный с ростом числа выпускаемых игр, но после количество выпускаемых игр становится меньше. С 2010 по 2016 годы происходит спад в индустрии видеоигр. Данная ситуация может быть связана с ростом популярности смартфонов.

### Шаг 3.2. Посмотрите, как менялись продажи по платформам. Выберите платформы с наибольшими суммарными продажами и постройте распределение по годам. За какой характерный срок появляются новые и исчезают старые платформы?

````python
top_10_platforms = (data.pivot_table(index='platform', values='total_sales', aggfunc='sum')\
                    .sort_values(by='total_sales',ascending=False).head(10).reset_index())
print(top_10_platforms) # отобразим таблицу c 10 платформами с самыми большими продажами
````

````python
for value in top_10_platforms['platform']: # для каждой из этих 10 платформ отобразим продажи по годам
    data.query('platform==@value').pivot_table(index='year_of_release', values='total_sales', aggfunc='sum')\
    .plot(kind='bar',
          xlabel='year_of_ralease',
          ylabel='total_sales')
    plt.title(f'{value}, Общие продажи по годам')
    plt.show()
````

Комментарий к графику по платформе DS: пплатформа DS была переименована в NES в 1985 году после старта продаж. А в 2004 вышла уже новая платформа с  названием DS.

Видим, что характерный срок от появления до исчезновения самых популярных платформ в разное время - от 11 до 12 лет. Это относится к платформам с наибольшими суммарными продажами: PS2, X360, PS3, Wii.
Платформа PS2 имеет наибольшие суммарные продажи.

### Шаг 3.4. Какие платформы лидируют по продажам, растут или падают? Выберите несколько потенциально прибыльных платформ.

````python
data_since_2012.pivot_table(index='platform', values='total_sales', aggfunc='sum').sort_values(by='total_sales',ascending=False)
````

Видим, что с 2012 года лидируют по продажам следующие платформы: PS4, PS3, X360.

````python
for platform in data_since_2012['platform'].unique():
    (
        data_since_2012
        .query('platform==@platform')
        .pivot_table(index=['year_of_release'], values=['total_sales'], aggfunc='sum')
        .plot(kind='bar',
             xlabel='year_of_release',
             ylabel='total_sales')
       )
    plt.title(f'{platform}, Общие продажи c 2012 года')
    plt.show()
````

Видим, что ни по одной из платформ продажи в 2016 году не увеличились относительно 2015 года, наблюдаем только падение продаж. Данная ситуация может быть связана с тем, что данные за 2016 год неполные.

````python
data_since_2012[data_since_2012['year_of_release'] == 2016]\
.pivot_table(index='platform', values='total_sales', aggfunc='sum')\
.sort_values(by='total_sales',ascending=False)
````

Самые прибыльные платформы на 2016 год - PS4, XOne, 3DS. Вероятно, что именно эти платформы будут потенциально прибыльны в 2017 году.

## Шаг 5. Проверка гипотезы

### Шаг 5.1. Проверьте гипотезу. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые?

```
H_0: Средний пользовательский рейтинг (user_score) платформы Xbox One = среднему пользовательскому рейтингу (user_score) платформы PC
H_1: Средний пользовательский рейтинг (user_score) платформы Xbox One ≠ среднему пользовательскому рейтингу (user_score) платформы PC

alpha = 0.05
```

````python
# results = вызов метода для проверки гипотезы
results = st.ttest_ind(data_since_2012.query('platform=="XOne" and user_score>0')['user_score'],
                       data_since_2012.query('platform=="PC" and user_score>0')['user_score'],
                       equal_var=False)
# alpha = задайте значение уровня значимости
alpha = 0.05
# вывод значения p-value на экран
print(results.pvalue)
# условный оператор с выводом строки с ответом
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
````

### Шаг 5.2. Проверьте гипотезу. Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

```
H_0: Средний пользовательский рейтинг (user_score) жанра Action = среднему пользовательскому рейтингу (user_score) жанра Sports
H_1: Средний пользовательский рейтинг (user_score) жанра Action ≠ среднему пользовательскому рейтингу (user_score) жанра Sports

alpha = 0.05
```

````python
# results = вызов метода для проверки гипотезы
results = st.ttest_ind(data_since_2012.query('genre=="Action" and user_score>0')['user_score'],
                       data_since_2012.query('genre=="Sports" and user_score>0')['user_score'],
                       equal_var=False)
# alpha = задайте значение уровня значимости
alpha = 0.05
# вывод значения p-value на экран
print(results.pvalue)
# условный оператор с выводом строки с ответом
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
````

Алгоритм проверки статистических гипотез всегда начинается с формулирования гипотез. Сначала формулируется нулевая гипотеза H_0. Исходя из H_0 формулируется альтернативная гипотеза H_1. Н_0 всегда формулируется так, чтобы использовать знак равенства. H_1 - гипотеза, противоположная по смыслу нулевой гипотизе.

Окончательное решение, какой порог(значение уровня значимости) считать достаточным,, всегда остаётся за аналитиком. Общепринятые пороговые значения - 5% и 1%.

**Вывод к шагу 5.**

Гипотезу о том, что средние пользовательские рейтинги платформ Xbox One и PC одинаковые, отвергнуть мы не можем.
Гипотезу о том, что средние пользовательские рейтинги жанров Action и Sports одинаковые, мы можем отвергнуть.

Однако платформы PS3, X360 к 2017 году становятся неактуальными из-за того, что прошло около 10 лет с момента их появления, а 10 лет - это характерный срок от появления до исчезновения популярных платформ.
- В лидерах по продажам в 2016 году стали платформы: PS4, XOne и 3DS. В 2017 году приоритет стоит отдать именно этим платформам.
- Связь между продажами и оценками экспертов для всех платформ является прямой, слабой степени выраженности (во всех случаях коэффициент около 0,3-0,4). Линейная связь между продажами и оценками пользователей для самых популярных платформ отсутствует, либо не является линейной. Учитывать оценки экспертов и оценки пользователей в выборе платформы не стоит.
- Жанры с самыми высокими медианными продажами - 'Shooter', Sports' и 'Platform'.  Жанры с наибольшим количеством проданных игр - 'Action', 'Shooter', 'Role-Playing', 'Sports'. <br>
__Примечание:__ в этом пункте platform (платформер)- жанр компьютерных игр, в которых основу игрового процесса составляют прыжки по платформам, лазанье по лестницам, сбор предметов, необходимых для победы над врагами или завершения уровня.
- Портрет пользователей для каждого региона различается. Для региона Северная Америка и Европа в 2017 году в первую очередь стоит сделать акцент на платформу 'PS4', затем на 'XOne' и только потом на '3DS'. В регионе 'JP' основной платформой является '3DS', ставку здесь стоит делать именно на эту платформу.
- Популярность жанров определяли по числу проданных копий в разных регионах. Для регионов Северная Америка и Европа самыми популярными жанрами являются: 'Action', 'Shooter', 'Sports', 'Role-Playing'. Только пятый по счёту самый популярный жанр отличается: в Северной Америке это 'Misc', а в Европу - 'Racing'.<br>
Для Японии видим иную картину. Там самыми популярными жанрами являются: 'Role-Playing', 'Action', 'Misc', 'Simulation', 'Fighting'.
- Рейтинг ESRB влияет на продажи в отдельном регионе. Для регионов Северная Америка и Европа самыми популярными рейтингами ESRB являются: 'M', 'E', 'unknown'.
Для Японии ситуация другая. Там самыми продаваемыми рейтингами ESRB являются: 'unknown', 'E', 'T'.
- Гипотезу о том, что средние пользовательские рейтинги платформ Xbox One и PC одинаковые, отвергнуть мы не можем. Гипотезу о том, что средние пользовательские рейтинги жанров Action и Sports одинаковые, мы можем отвергнуть.
